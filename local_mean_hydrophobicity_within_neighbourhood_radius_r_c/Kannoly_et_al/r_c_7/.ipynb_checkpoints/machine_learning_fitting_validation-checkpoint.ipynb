# Calculating the correlation between MFPT and mean hydrophobicity

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator  # Import AutoMinorLocator
from scipy.stats import pearsonr

# Read data from file
data_file = "MFPT_CV2_Average_Hydrophobicity.txt"

mutants = []
mfpt = []
stderr_mfpt = []  # Adding a list for standard error of MFPT
mean_hydrophobicity = []

# Corrected mutant names
name_list_corrected = {
    "WT": "WT",
    "S105": "S105",
    "JJD246": "H7D",
    "JJD248": "F94C",
    "JJD251": "A99V",
    "JJD253": "L10M",
    "JJD388": "L25V/N37H",
    "JJD390": "A11G/Y31H",
    "JJD391": "A16G/K92Q",
    "JJD404": "I21V",
    "JJD405": "V45G",
    "JJD411": "D85G",
    "JJD413": "I87L",
    "JJD414": "L90I",
    "JJD415": "I91T",
    "JJD426": "G38S",
    "JJD428": "G39S",
    "JJD432": "S89W",
    "JJD434": "D8G",
    "JJD436": "K92N",
}

with open(data_file, 'r') as file:
    next(file)  # Skip the header line
    for line in file:
        parts = line.strip().split()
        mutant = parts[0]
        mfpt_val = float(parts[1])
        stderr_mfpt_val = float(parts[2])  # Extracting standard error of MFPT
        mean_hydrophobicity_val = float(parts[5])

        # Filter based on condition (MFPT < 55)
        #if mfpt_val < 200:
        if mutant not in ["JJD405", "JJD413"] and mfpt_val < 200:
            mutants.append(name_list_corrected[mutant])
            mfpt.append(mfpt_val)
            stderr_mfpt.append(stderr_mfpt_val)  # Appending standard error of MFPT
            mean_hydrophobicity.append(mean_hydrophobicity_val)

# Calculate Pearson correlation coefficient
correlation_coefficient, p_value = pearsonr(mean_hydrophobicity, mfpt)

# Calculate slope and intercept
slope, intercept = np.polyfit(mean_hydrophobicity, mfpt, 1)

# Print slope, intercept, and correlation coefficient
print("Slope:", slope)
print("Intercept:", intercept)
print("Correlation coefficient:", correlation_coefficient)


# Plot
plt.figure(figsize=(8, 7))
plt.errorbar(mean_hydrophobicity, mfpt, yerr=stderr_mfpt, fmt='o', color='green', capsize=5, markersize=12)  # Include error bars

# Draw correlation line
plt.plot(np.unique(mean_hydrophobicity), np.poly1d(np.polyfit(mean_hydrophobicity, mfpt, 1))(np.unique(mean_hydrophobicity)), color='red', linewidth=3)

# Annotate each point with mutant name
for i, mutant in enumerate(mutants):
    plt.annotate(mutant, (mean_hydrophobicity[i], mfpt[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=15)

plt.axis([-0.22, 0.57, 0, 180])
plt.xlabel('Mean Hydrophobicity', fontsize=25, color='black', fontweight='bold')
plt.ylabel('Lysis Time (min)', fontsize=25, color='black', fontweight='bold')
#plt.title(f'Pearson Correlation Coefficient: {correlation_coefficient:.2f}', fontsize=20, color='red', fontweight='bold')
#plt.title(f'$R$ = 7', fontsize=25, color='black')

# Display p-value
#plt.text(0.94, 0.92, f'p-value: {p_value:.2f}', transform=plt.gca().transAxes, fontsize=18, ha='right', color='black', style='italic')
plt.text(0.95, 0.92, f'r = {correlation_coefficient:.2f}', transform=plt.gca().transAxes, fontsize=20, ha='right', color='black', style='italic')
plt.text(0.95, 0.84, f'p-value = {p_value:.2f}', transform=plt.gca().transAxes, fontsize=20, ha='right', color='black', style='italic')

plt.xticks(fontsize=20)  # Indicate the size of x-ticks
plt.yticks(fontsize=20)  # Indicate the size of y-ticks
plt.tick_params(axis='x', which='both', bottom=True, top=True, direction='in')
plt.tick_params(axis='y', which='both', left=True, right=True, direction='in')
plt.minorticks_on()
plt.gca().xaxis.set_minor_locator(AutoMinorLocator(2))
plt.gca().yaxis.set_minor_locator(AutoMinorLocator(2))
plt.rcParams["axes.edgecolor"]="black"
plt.rcParams["axes.linewidth"]=1

plt.grid(False)
plt.tight_layout()
plt.savefig('Hydrophobicity_vs_MFPT.pdf', bbox_inches='tight')
plt.show()
```

# Training the data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Define function to calculate slope and intercept using np.polyfit
def calculate_slope_intercept(X_train, y_train):
    slope, intercept = np.polyfit(X_train, y_train, 1)
    return slope, intercept

# Define function to calculate confidence interval
def calculate_confidence_interval(data, confidence_level=0.95):
    lower_percentile = (1 - confidence_level) / 2 * 100
    upper_percentile = 100 - lower_percentile
    lower_bound = np.percentile(data, lower_percentile)
    upper_bound = np.percentile(data, upper_percentile)
    return lower_bound, upper_bound

# Parameters
training_size = 0.25
CV = 10000  # Cross validation
confidence_level = 0.95  # 95% confidence interval

# Initialize lists to store slope and intercept values
slope_list = []
intercept_list = []

# Bootstrap resampling
for i in range(CV):
    # Splitting data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(mean_hydrophobicity, mfpt, test_size=training_size, random_state=None)
    
    # Calculating slope and intercept for this sample
    slope, intercept = calculate_slope_intercept(X_train, y_train)
    
    # Append slope and intercept to lists
    slope_list.append(slope)
    intercept_list.append(intercept)

# Calculate confidence interval for slope
slope_lower, slope_upper = calculate_confidence_interval(slope_list, confidence_level)

# Calculate confidence interval for intercept
intercept_lower, intercept_upper = calculate_confidence_interval(intercept_list, confidence_level)

# Print confidence intervals
print("Confidence interval for slope: [{:.4f}, {:.4f}]".format(slope_lower, slope_upper))
print("Confidence interval for intercept: [{:.4f}, {:.4f}]".format(intercept_lower, intercept_upper))

# Filter slope values within confidence interval
filtered_slope_list = [slope for slope in slope_list if slope >= slope_lower and slope <= slope_upper]

# Filter intercept values within confidence interval
filtered_intercept_list = [intercept for intercept in intercept_list if intercept >= intercept_lower and intercept <= intercept_upper]

# Calculate average slope within confidence interval
average_slope_within_ci = np.mean(filtered_slope_list)

# Calculate average intercept within confidence interval
average_intercept_within_ci = np.mean(filtered_intercept_list)

print("Average slope within confidence interval: {:.4f}".format(average_slope_within_ci))
print("Average intercept within confidence interval: {:.4f}".format(average_intercept_within_ci))


# Plot histograms
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.hist(slope_list, bins='auto', color='skyblue', edgecolor='black')
plt.title('Histogram of Slopes')
plt.xlabel('Slope')
plt.ylabel('Frequency')
plt.grid(False)

plt.subplot(1, 2, 2)
plt.hist(intercept_list, bins='auto', color='salmon', edgecolor='black')
plt.title('Histogram of Intercepts')
plt.xlabel('Intercept')
plt.ylabel('Frequency')
plt.grid(False)

plt.tight_layout()
plt.show()
```

# Creating a dictionay for predicted and actual lysis time

```python
mfpt_dict={}

for i, j in zip(X_test, y_test):
    mfpt_predicted = average_slope_within_ci * i + average_intercept_within_ci
    mfpt_predicted_list.append(mfpt_predicted)
    mfpt_actual_list.append(j)
    mfpt_dict[mfpt_predicted]=j
    print('predicted mfpt='+ str(mfpt_predicted),'  actual mfpt='+str(j))
    #print(np.abs(mfpt_predicted-j))
```

```python
mfpt_predicted_list = list(mfpt_dict.keys())
mfpt_actual_list = list(mfpt_dict.values())
mfpt_predicted_list
```

# Calculating R-squared (Coefficient of Determination)

```python
from sklearn.metrics import r2_score

mean_actual_mfpt = np.mean(mfpt_actual_list)
sumdiff=0
sumavgdiff=0
for i in range(len(mfpt_predicted_list)):
    sumdiff += ((mfpt_predicted_list[i] - mfpt_actual_list[i]) ** 2)
    sumavgdiff += ((mfpt_actual_list[i] - mean_actual_mfpt) ** 2)
    
R2sq = 1 - (sumdiff/sumavgdiff) # calculated using mathematical formula
r_squared = r2_score(mfpt_actual_list, mfpt_predicted_list) # calculated using in-built function
print(R2sq)
print(r_squared)    
```
