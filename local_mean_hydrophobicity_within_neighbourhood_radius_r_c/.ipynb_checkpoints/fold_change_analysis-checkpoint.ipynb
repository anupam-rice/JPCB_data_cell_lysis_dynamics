# Calculate hydrophobicity of a protein sequence (Kyte-Doolittle Scale)

```python
# Define the Kyte-Doolittle hydrophobicity scale as a dictionary
hydrophobicity_scale = {
    'A': 1.8, 'R': -4.5, 'N': -3.5, 'D': -3.5, 'C': 2.5,
    'Q': -3.5, 'E': -3.5, 'G': -0.4, 'H': -3.2, 'I': 4.5,
    'L': 3.8, 'K': -3.9, 'M': 1.9, 'F': 2.8, 'P': -1.6,
    'S': -0.8, 'T': -0.7, 'W': -0.9, 'Y': -1.3, 'V': 4.2
}

def calculate_hydrophobicity(protein_sequence):
    """
    Calculates the hydrophobicity of a protein sequence using the Kyte-Doolittle scale.
    
    Args:
    - protein_sequence (str): A string representing the protein sequence
    
    Returns:
    - hydrophobicity (float): The calculated hydrophobicity of the protein sequence
    """
    hydrophobicity = 0.0
    for aa in protein_sequence:
        if aa in hydrophobicity_scale:
            hydrophobicity += hydrophobicity_scale[aa]
    return hydrophobicity

# Example usage:
protein_sequence = "DDFLNSAMSSLYSGWSSFTTGASKFASAAKEGATK"
hydrophobicity = calculate_hydrophobicity(protein_sequence)
print("Hydrophobicity of the protein sequence:", hydrophobicity)
```

# Calculate hydrophobicity of a protein sequence (Wimley-White Scale)
Wimley, W. C.; White, S. H. Experimentally determined hydrophobicity scale for proteins at membrane interfaces. Nat. Struct. Biol. 1996, 3, 842â€“848.

```python
# Define the Wimley-White hydrophobicity scale as a dictionary
hydrophobicity_scale = {
    'A': 0.17, 'R': 0.81, 'N': 0.42, 'D': 1.23, 'C': -0.24,
    'Q': 0.58, 'E': 2.02, 'G': 0.01, 'H': 0.96, 'I': -0.31,
    'L': -0.56, 'K': 0.99, 'M': -0.23, 'F': -1.13, 'P': 0.45,
    'S': 0.13, 'T': 0.14, 'W': -1.85, 'Y': -0.94, 'V': 0.07
}


def calculate_hydrophobicity_and_mean(protein_sequence):
    """
    Calculates the hydrophobicity of a protein sequence using the Wimley-White scale
    and also calculates the mean hydrophobicity.
    
    Args:
    - protein_sequence (str): A string representing the protein sequence
    
    Returns:
    - hydrophobicity (float): The calculated hydrophobicity of the protein sequence
    - mean_hydrophobicity (float): The mean hydrophobicity of the protein sequence
    """
    hydrophobicity = 0.0
    num_aa = 0
    for aa in protein_sequence:
        if aa in hydrophobicity_scale:
            hydrophobicity += hydrophobicity_scale[aa]
            num_aa += 1
    if num_aa != 0:
        mean_hydrophobicity = hydrophobicity / num_aa
    else:
        mean_hydrophobicity = None
    return hydrophobicity, mean_hydrophobicity

# Example usage:
protein_sequence = "MPEKHD"
hydrophobicity, mean_hydrophobicity = calculate_hydrophobicity_and_mean(protein_sequence)
print("Hydrophobicity of the protein sequence (Wimley-White scale):", hydrophobicity)
print("Mean hydrophobicity of the protein sequence:", mean_hydrophobicity)
```

# Correlation Between Mean Hydrophobicity and Mean Frustration for all Holin Mutants of iScience Paper

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

# Corrected mutant names
name_list_corrected = {
    "WT": "WT",
    "S105": "S105",
    "JJD246": "H7D",
    "JJD248": "F94C",
    "JJD251": "A99V",
    "JJD253": "L10M",
    "JJD388": "L25V/N37H",
    "JJD390": "A11G/Y31H",
    "JJD391": "A16G/K92Q",
    "JJD404": "I21V",
    "JJD405": "V45G",
    "JJD411": "D85G",
    "JJD413": "I87L",
    "JJD414": "L90I",
    "JJD415": "I91T",
    "JJD426": "G38S",
    "JJD428": "G39S",
    "JJD432": "S89W",
    "JJD434": "D8G",
    "JJD436": "K92N",
}

# Read data from file
data_file = "Mean_Hydrophobicity_and_Mean_Frustration_of_all_Holin_Mutants_Subdomains_iScience.txt"
with open(data_file, 'r') as file:
    next(file)  # Skip header
    lines = file.readlines()

domains = []
mean_hydrophobicity = []
mean_frustration = []
stderr_frustration = []

for line in lines:
    parts = line.split()
    domains.append(parts[0])
    mean_hydrophobicity.append(float(parts[3]))
    mean_frustration.append(float(parts[1]))
    stderr_frustration.append(float(parts[2]))

# Replace mutant names in domains list
for i, domain in enumerate(domains):
    if domain in name_list_corrected:
        domains[i] = name_list_corrected[domain]

# Plot
plt.figure(figsize=(8, 6))
plt.errorbar(mean_hydrophobicity, mean_frustration, yerr=stderr_frustration, fmt='o', color='blue', markersize=10)  # Include error bars

# Annotate each point with domain name
for i, domain in enumerate(domains):
    plt.annotate(domain, (mean_hydrophobicity[i], mean_frustration[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=12)

# Calculate Pearson correlation coefficient
correlation_coefficient, p_value = pearsonr(mean_hydrophobicity, mean_frustration)
plt.title(f'Pearson Correlation Coefficient: {correlation_coefficient:.2f}', fontsize=20, color='red', fontweight='bold')

# Plot the correlation coefficient as a red line
plt.plot(np.unique(mean_hydrophobicity), np.poly1d(np.polyfit(mean_hydrophobicity, mean_frustration, 1))(np.unique(mean_hydrophobicity)), color='red')

# Labeling axes
plt.xlabel('Mean Hydrophobicity', fontsize=20, color='black', fontweight='bold')
plt.ylabel('Mean Frustration', fontsize=20, color='black', fontweight='bold')
# Set the size of x-ticks and y-ticks
plt.xticks(fontsize=20)  # Indicate the size of x-ticks
plt.yticks(fontsize=20)  # Indicate the size of y-ticks
# Mirror the x-ticks and y-ticks
plt.tick_params(axis='x', which='both', bottom=True, top=True, direction='in')
plt.tick_params(axis='y', which='both', left=True, right=True, direction='in')
plt.minorticks_on()
# Add one minor tick in between each pair of major ticks on x-axis and y-axis
plt.gca().xaxis.set_minor_locator(AutoMinorLocator(2))
plt.gca().yaxis.set_minor_locator(AutoMinorLocator(1))

# Display p-value
plt.text(0.45, 0.1, f'p-value: {p_value:.7f}', transform=plt.gca().transAxes, fontsize=18, ha='right', color='black', style='italic')


# Display plot
plt.grid(False)
plt.show()
```

# Correlation between mean hydrophobicity and average holin number

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

# Read data from file
data_file = "Mean_Hydrophibicity_and_Holin_Numbers.txt"
mutants = []
mean_hydrophobicity = []
holin_numbers = []

with open(data_file, 'r') as file:
    next(file)  # Skip header
    for line in file:
        mutant, hydrophobicity, holin_number = line.strip().split()
        mutants.append(mutant)
        mean_hydrophobicity.append(float(hydrophobicity))
        holin_numbers.append(float(holin_number))

# Corrected mutant names
name_list_corrected = {
    "WT": "WT",
    "S105": "S105",
    "JJD246": "H7D",
    "JJD248": "F94C",
    "JJD251": "A99V",
    "JJD253": "L10M",
    "JJD388": "L25V/N37H",
    "JJD390": "A11G/Y31H",
    "JJD391": "A16G/K92Q",
    "JJD404": "I21V",
    "JJD405": "V45G",
    "JJD411": "D85G",
    "JJD413": "I87L",
    "JJD414": "L90I",
    "JJD415": "I91T",
    "JJD426": "G38S",
    "JJD428": "G39S",
    "JJD432": "S89W",
    "JJD434": "D8G",
    "JJD436": "K92N",
}

# Update mutant names
mutants_corrected = [name_list_corrected.get(mutant, mutant) for mutant in mutants]

# Calculate Pearson correlation coefficient
correlation_coefficient, _ = pearsonr(mean_hydrophobicity, holin_numbers)

# Print correlation coefficient
print("Pearson correlation coefficient:", correlation_coefficient)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(mean_hydrophobicity, holin_numbers, color='blue')

# Add corrected mutant names
for i, mutant in enumerate(mutants_corrected):
    plt.text(mean_hydrophobicity[i], holin_numbers[i], mutant, fontsize=9)

# Add regression line
m, b = np.polyfit(mean_hydrophobicity, holin_numbers, 1)
plt.plot(mean_hydrophobicity, m*np.array(mean_hydrophobicity) + b, color='red')

# Labeling and title
plt.xlabel('Mean Hydrophobicity')
plt.ylabel('Average Holin Numbers')
plt.title('Correlation between Mean Hydrophobicity and Average Holin Numbers\nCorrelation Coefficient: {:.2f}'.format(correlation_coefficient))

# Show plot
plt.grid(False)
plt.show()
```
