# Calculate Mean Hydrophobicity of a Protein Sequence (Wimley-White Scale)
Wimley, W. C.; White, S. H. Experimentally determined hydrophobicity scale for proteins at membrane interfaces. Nat. Struct. Biol. 1996, 3, 842â€“848.

```python
# Define the Wimley-White hydrophobicity scale as a dictionary
hydrophobicity_scale = {
    'A': 0.17, 'R': 0.81, 'N': 0.42, 'D': 1.23, 'C': -0.24,
    'Q': 0.58, 'E': 2.02, 'G': 0.01, 'H': 0.96, 'I': -0.31,
    'L': -0.56, 'K': 0.99, 'M': -0.23, 'F': -1.13, 'P': 0.45,
    'S': 0.13, 'T': 0.14, 'W': -1.85, 'Y': -0.94, 'V': 0.07
}


def calculate_hydrophobicity_and_mean(protein_sequence):
    """
    Calculates the hydrophobicity of a protein sequence using the Wimley-White scale
    and also calculates the mean hydrophobicity.
    
    Args:
    - protein_sequence (str): A string representing the protein sequence
    
    Returns:
    - hydrophobicity (float): The calculated hydrophobicity of the protein sequence
    - mean_hydrophobicity (float): The mean hydrophobicity of the protein sequence
    """
    hydrophobicity = 0.0
    num_aa = 0
    for aa in protein_sequence:
        if aa in hydrophobicity_scale:
            hydrophobicity += hydrophobicity_scale[aa]
            num_aa += 1
    if num_aa != 0:
        mean_hydrophobicity = hydrophobicity / num_aa
    else:
        mean_hydrophobicity = None
    return hydrophobicity, mean_hydrophobicity

# Example usage:
protein_sequence = "TDSIGWLIKRF"
hydrophobicity, mean_hydrophobicity = calculate_hydrophobicity_and_mean(protein_sequence)
print("Hydrophobicity of the protein sequence (Wimley-White scale):", hydrophobicity)
print("Mean hydrophobicity of the protein sequence:", mean_hydrophobicity)
```

# Calculate Mean Hydrophobicity for Multiple Sequences

```python
# Define the Wimley-White hydrophobicity scale as a dictionary
hydrophobicity_scale = {
    'A': 0.17, 'R': 0.81, 'N': 0.42, 'D': 1.23, 'C': -0.24,
    'Q': 0.58, 'E': 2.02, 'G': 0.01, 'H': 0.96, 'I': -0.31,
    'L': -0.56, 'K': 0.99, 'M': -0.23, 'F': -1.13, 'P': 0.45,
    'S': 0.13, 'T': 0.14, 'W': -1.85, 'Y': -0.94, 'V': 0.07
}

def calculate_hydrophobicity_and_mean(protein_sequence):
    hydrophobicity = 0.0
    num_aa = 0
    for aa in protein_sequence:
        if aa in hydrophobicity_scale:
            hydrophobicity += hydrophobicity_scale[aa]
            num_aa += 1
    if num_aa != 0:
        mean_hydrophobicity = hydrophobicity / num_aa
    else:
        mean_hydrophobicity = None
    return hydrophobicity, mean_hydrophobicity

# Read sequences from file
with open("holin_virtually_mutated_seq.txt", "r") as file:
    sequences = [line.strip() for line in file if not line.startswith(">")]

# Calculate and print average hydrophobicity for each sequence
print(f"Avg_Hydrophobicity")
for sequence in sequences:
    hydrophobicity, _ = calculate_hydrophobicity_and_mean(sequence)
    num_aa = len(sequence)
    average_hydrophobicity = hydrophobicity / num_aa
    #print(f"Average hydrophobicity for sequence '{sequence}': {average_hydrophobicity}")
    print(f"{average_hydrophobicity:.3f}")
    
```

# Predicting Lysis Time based on Mean Hydrophobicity Values

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator

# Read data from file
data_file = "Average_Hydrophobicity.txt"
average_slope = -143.40
average_intercept = 78.56

mutants = []
mean_hydrophobicity = []

with open(data_file, 'r') as file:
    next(file)  # Skip the header line
    for line in file:
        parts = line.strip().split()
        mutants.append(parts[0])
        mean_hydrophobicity.append(float(parts[1]))

# Sort the data by mean hydrophobicity
sorted_indices = sorted(range(len(mean_hydrophobicity)), key=lambda k: mean_hydrophobicity[k])
mutants = [mutants[i] for i in sorted_indices]
mean_hydrophobicity = [mean_hydrophobicity[i] for i in sorted_indices]

# Plot
plt.figure(figsize=(18, 8))
for mutant, hydrophobicity in zip(mutants, mean_hydrophobicity):
    mfpt_predicted = average_slope * hydrophobicity + average_intercept
    if mutant == "D85G":
        plt.bar(mutant, mfpt_predicted, color='red', alpha=1.0)
    else:
        plt.bar(mutant, mfpt_predicted, color='royalblue', alpha=1.0)

plt.xlabel('Holin Mutants at Position D85', fontsize=30, color='black', fontweight='bold')
plt.ylabel('Predicted Lysis Time (min)', fontsize=30, color='black', fontweight='bold')
plt.xticks(rotation=45, fontsize=25)
plt.yticks(fontsize=25)
plt.tick_params(axis='x', which='both', bottom=True, top=True, direction='in', labelsize=25)
plt.tick_params(axis='y', which='both', left=True, right=True, direction='in', labelsize=25)
plt.rcParams["axes.edgecolor"] = "black"
plt.rcParams["axes.linewidth"] = 1
plt.minorticks_on()
plt.gca().xaxis.set_minor_locator(AutoMinorLocator(1))
plt.gca().yaxis.set_minor_locator(AutoMinorLocator(2))
plt.xlim(-0.8,18.8)

plt.tight_layout()
plt.show()
```
