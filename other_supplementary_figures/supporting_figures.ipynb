# Correlation between Residue Volume and Mean Lysis Timing

## Bigelow, C. C. On the average hydrophobicity of proteins and the relation between it and protein structure. J Theor Biol. 1967, 16, 187-211.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator  # Import AutoMinorLocator
from scipy.stats import pearsonr

# Read data from file
data_file = "residue_volume_vs_MFPT_Bigelow.txt"

mutants = []
mfpt = []
stderr_mfpt = []  # Adding a list for standard error of MFPT
mean_volume = []

# Corrected mutant names
name_list_corrected = {
    "H7D": "H7D",
    "F94C": "F94C",
    "A99V": "A99V",
    "L10M": "L10M",
    "I21V": "I21V",
    "V45G": "V45G",
    "D85G": "D85G",
    "I87L": "I87L",
    "L90I": "L90I",
    "I91T": "I91T",
    "G38S": "G38S",
    "G39S": "G39S",
    "S89W": "S89W",
    "D8G": "D8G",
    "K92N": "K92N",
}

with open(data_file, 'r') as file:
    next(file)  # Skip the header line
    for line in file:
        parts = line.strip().split()
        mutant = parts[0]
        mean_volume_val = float(parts[1])
        mfpt_val = float(parts[2])
        stderr_mfpt_val = float(parts[3])  # Extracting standard error of MFPT

        # Exclude "I87L" mutant and filter based on condition (MFPT < 55)
        if mutant not in ["V45G", "S89W"]:
        #if mfpt_val < 200:
            mutants.append(name_list_corrected[mutant])
            mfpt.append(mfpt_val)
            stderr_mfpt.append(stderr_mfpt_val)  # Appending standard error of MFPT
            mean_volume.append(mean_volume_val)

# Calculate Pearson correlation coefficient
correlation_coefficient, p_value = pearsonr(mean_volume, mfpt)

# Plot
plt.figure(figsize=(8, 8))
#plt.figure(figsize=(8.2, 7.5))
plt.errorbar(mean_volume, mfpt, yerr=stderr_mfpt, fmt='o', color='green', capsize=5, markersize=14)  # Include error bars

# Draw correlation line
plt.plot(np.unique(mean_volume), np.poly1d(np.polyfit(mean_volume, mfpt, 1))(np.unique(mean_volume)), color='red', linewidth=4)

# Annotate each point with mutant name
#for i, mutant in enumerate(mutants):
#    plt.annotate(mutant, (mean_volume[i], mfpt[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=16)

for i, mutant in enumerate(mutants):
    if mutant == "D85G":
        plt.annotate("D85G", (mean_volume[i], mfpt[i]), textcoords="offset points", xytext=(0,15), ha='center', fontsize=18)
    elif mutant == "G38S":
        plt.annotate("G38S", (mean_volume[i], mfpt[i]), textcoords="offset points", xytext=(0,12), ha='center', fontsize=18)
    elif mutant == "I21V":
        plt.annotate("I21V", (mean_volume[i], mfpt[i]), textcoords="offset points", xytext=(0,-22), ha='center', fontsize=18)
    elif mutant == "I87L":
        plt.annotate("I87L", (mean_volume[i], mfpt[i]), textcoords="offset points", xytext=(0,-24), ha='center', fontsize=18)
    elif mutant == "L10M":
        plt.annotate("L10M", (mean_volume[i], mfpt[i]), textcoords="offset points", xytext=(-15,-22), ha='center', fontsize=18)
    else:
        plt.annotate(mutant, (mean_volume[i], mfpt[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=18)

    
plt.axis([28, 110, 0, 180])
#plt.xticks([60,80,100,120,140,160,180])
plt.xlabel('Residue Volume ()', fontsize=25, color='black', fontweight='bold')
plt.ylabel('Mean Lysis Time (min)', fontsize=25, color='black', fontweight='bold')


plt.tick_params(axis='x', which='major', bottom=True, top=True, direction='in', labelsize=22, width=2, length=7)
plt.tick_params(axis='x', which='minor', bottom=True, top=True, direction='in', labelsize=22, width=2, length=4)
plt.tick_params(axis='y', which='major', left=True, right=True, direction='in', labelsize=22, width=2, length=7)
plt.tick_params(axis='y', which='minor', left=True, right=True, direction='in', labelsize=22, width=2, length=4)
plt.minorticks_on()
plt.gca().xaxis.set_minor_locator(AutoMinorLocator(2))
plt.gca().yaxis.set_minor_locator(AutoMinorLocator(2))
plt.rcParams["axes.edgecolor"]="black"
plt.rcParams["axes.linewidth"]=1.5

# Display p-value
#plt.text(-0.165, 1.1, "A", transform=plt.gca().transAxes, fontsize=35, fontweight='bold', va='top')
plt.text(0.93, 0.92, f'r = ${correlation_coefficient:.2f}$', transform=plt.gca().transAxes, fontsize=22, ha='right', color='black', style='italic')
plt.text(0.93, 0.84, f'p-value = {p_value:.3f}', transform=plt.gca().transAxes, fontsize=22, ha='right', color='black', style='italic')

plt.grid(False)
plt.tight_layout()
#plt.savefig('ResidueVol_vs_MFPT.pdf', bbox_inches='tight')
plt.show()
```

# Correlation between Polarizability and Mean Lysis Timing

## Charton, M.  and Charton, B. I. The structural dependence of amino acid hydrophobicity parameters. J Theor Biol. 1982, 99, 629-644.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import AutoMinorLocator  # Import AutoMinorLocator
from scipy.stats import pearsonr

# Read data from file
data_file = "polarizability_vs_MFPT_Charton.txt"

mutants = []
mfpt = []
stderr_mfpt = []  # Adding a list for standard error of MFPT
polarizability = []

# Corrected mutant names
name_list_corrected = {
    "H7D": "H7D",
    "F94C": "F94C",
    "A99V": "A99V",
    "L10M": "L10M",
    "I21V": "I21V",
    "V45G": "V45G",
    "D85G": "D85G",
    "I87L": "I87L",
    "L90I": "L90I",
    "I91T": "I91T",
    "G38S": "G38S",
    "G39S": "G39S",
    "S89W": "S89W",
    "D8G": "D8G",
    "K92N": "K92N",
}

with open(data_file, 'r') as file:
    next(file)  # Skip the header line
    for line in file:
        parts = line.strip().split()
        mutant = parts[0]
        polarizability_val = float(parts[1])
        mfpt_val = float(parts[2])
        stderr_mfpt_val = float(parts[3])  # Extracting standard error of MFPT

        # Exclude "I87L" mutant and filter based on condition (MFPT < 55)
        if mutant not in ["V45G", "S89W"]:
        #if mfpt_val < 200:
            mutants.append(name_list_corrected[mutant])
            mfpt.append(mfpt_val)
            stderr_mfpt.append(stderr_mfpt_val)  # Appending standard error of MFPT
            polarizability.append(polarizability_val)

# Calculate Pearson correlation coefficient
correlation_coefficient, p_value = pearsonr(polarizability, mfpt)

# Plot
plt.figure(figsize=(8, 8))
#plt.figure(figsize=(8.2, 7.5))
plt.errorbar(polarizability, mfpt, yerr=stderr_mfpt, fmt='o', color='green', capsize=5, markersize=14)  # Include error bars

# Draw correlation line
plt.plot(np.unique(polarizability), np.poly1d(np.polyfit(polarizability, mfpt, 1))(np.unique(polarizability)), color='red', linewidth=4)

# Annotate each point with mutant name
#for i, mutant in enumerate(mutants):
#    plt.annotate(mutant, (polarizability[i], mfpt[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=16)

for i, mutant in enumerate(mutants):
    if mutant == "D85G":
        plt.annotate("D85G", (polarizability[i], mfpt[i]), textcoords="offset points", xytext=(0,15), ha='center', fontsize=18)
    elif mutant == "G38S":
        plt.annotate("G38S", (polarizability[i], mfpt[i]), textcoords="offset points", xytext=(0,12), ha='center', fontsize=18)
    elif mutant == "I21V":
        plt.annotate("I21V", (polarizability[i], mfpt[i]), textcoords="offset points", xytext=(0,-22), ha='center', fontsize=18)
    elif mutant == "I87L":
        plt.annotate("I87L", (polarizability[i], mfpt[i]), textcoords="offset points", xytext=(0,-24), ha='center', fontsize=18)
    else:
        plt.annotate(mutant, (polarizability[i], mfpt[i]), textcoords="offset points", xytext=(0,10), ha='center', fontsize=18)

    
plt.axis([-0.025, 0.245, 0, 180])
#plt.xticks([60,80,100,120,140,160,180])
plt.xlabel('Polarizability', fontsize=25, color='black', fontweight='bold')
plt.ylabel('Mean Lysis Time (min)', fontsize=25, color='black', fontweight='bold')


plt.tick_params(axis='x', which='major', bottom=True, top=True, direction='in', labelsize=22, width=2, length=7)
plt.tick_params(axis='x', which='minor', bottom=True, top=True, direction='in', labelsize=22, width=2, length=4)
plt.tick_params(axis='y', which='major', left=True, right=True, direction='in', labelsize=22, width=2, length=7)
plt.tick_params(axis='y', which='minor', left=True, right=True, direction='in', labelsize=22, width=2, length=4)
plt.minorticks_on()
plt.gca().xaxis.set_minor_locator(AutoMinorLocator(2))
plt.gca().yaxis.set_minor_locator(AutoMinorLocator(2))
plt.rcParams["axes.edgecolor"]="black"
plt.rcParams["axes.linewidth"]=1.5

# Display p-value
#plt.text(-0.165, 1.1, "C", transform=plt.gca().transAxes, fontsize=35, fontweight='bold', va='top')
plt.text(0.93, 0.92, f'r = ${correlation_coefficient:.2f}$', transform=plt.gca().transAxes, fontsize=22, ha='right', color='black', style='italic')
plt.text(0.93, 0.84, f'p-value = {p_value:.3f}', transform=plt.gca().transAxes, fontsize=22, ha='right', color='black', style='italic')

plt.grid(False)
plt.tight_layout()
#plt.savefig('Polarizability_vs_MFPT.pdf', bbox_inches='tight')
plt.show()
```
